```python
import pandas as pd
```

```python
inflacion_df = pd.read_html(r"C:\Users\justi\ETL-Project\inflacion.html")[0]
inflacion_df
```

```python
inflacion_df = inflacion_df.rename(columns={'Unnamed: 0': 'Date', 'Unnamed: 1': 'Inflation Rate'})
inflacion_df.head()
```

```python
# new data frame with split value columns 
new = inflacion_df["Date"].str.split(" ", n = 1, expand = True) 
  
# making separate first name column from new data frame 
inflacion_df["Month"]= new[0] 
  
# making separate last name column from new data frame 
inflacion_df["Year"]= new[1] 
  
# Dropping old Name columns 
inflacion_df.drop(columns =["Date"], inplace = True) 
  
# df display 
inflacion_df.head(12)
```

```python
inflacion_df["Month"].replace("Ene", "enero", inplace=True)
inflacion_df["Month"].replace("Feb", "febrero", inplace=True)
inflacion_df["Month"].replace("Mar", "marzo", inplace=True)
inflacion_df["Month"].replace("Abr", "abril", inplace=True)
inflacion_df["Month"].replace("May", "mayo", inplace=True)
inflacion_df["Month"].replace("Jun", "junio", inplace=True)
inflacion_df["Month"].replace("Jul", "julio", inplace=True)
inflacion_df["Month"].replace("Ago", "agosto", inplace=True)
inflacion_df["Month"].replace("Sep", "septiembre", inplace=True)
inflacion_df["Month"].replace("Oct", "octubre", inplace=True)
inflacion_df["Month"].replace("Nov", "noviembre", inplace=True)
inflacion_df["Month"].replace("Dic", "diciembre", inplace=True)
inflacion_df.head(12)
```

```python
#inflacion_df.insert(0,"cliente", "NA")
inflacion_df.insert(0,"cliente", "NA")
inflacion_df.insert(0,"estado", "NA")
inflacion_df.insert(0,"modalidad", "NA")
inflacion_df.insert(0,"esquema", "NA")
inflacion_df.head(12)
```

```python
inflacion_ordered=inflacion_df.loc[:,['Year','Month','estado','cliente','modalidad','Inflation Rate']]
inflacion_final=inflacion_ordered.rename(columns={"Year":"ano","Month":"mes", "Inflation Rate":"inflacion"})
inflacion_final.head()
```

```python
inflacion_final_df = inflacion_final.round({'inflacion': 2})
inflacion_final_df.head()
```
