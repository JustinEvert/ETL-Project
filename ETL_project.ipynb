```python
import pandas as pd
import numpy as np
import requests
```

```python
pages = np.arange(1,264)
data = []
for x in pages:
    query = f"https://api.datos.gob.mx/v1/fonden.reconstruccion/?page={x}"
    response = requests.get(query)
    print(f"Processing page {x} {response.status_code}: {query}")
    json = response.json()
    data.append(json)
```

```python
results = []
for x in data:
    for y in x['results']:
        results.append(y)
```

```python
results[0]
```

```python
id = []
edo = []
mun = []
evt = []
fec = []
apy = []
acc = []

for x in results:
    id.append(x['_id'])
    edo.append(x['nom_ent'])
    mun.append(x['nom_mun'])
    evt.append(x['evento'])
    fec.append(x['fecha'])
    apy.append(x['tipo_de_apoyo'])
    acc.append(x['accion'])
```

```python
df1 = pd.DataFrame({
"id": id,
"Estado": edo,
"Municipio": mun,
"Evento": evt,
"Fecha": fec,
"Tipo de Apoyo": apy,
"Accion": acc
})
df1.head()
```

```python
df1[["Año", "Mes", "Dia"]] = df1["Fecha"].str.split("-", expand = True)
df1.head()
```

```python
session = Session(db)
```

```python
from sqlalchemy import MetaData
meta = MetaData()
meta.reflect(bind=db)
```

```python
consolidated = meta.tables['consolidated']
```

```python
fonden = meta.tables['fonden']
```

```python
query1 = session.query(consolidated.columns.estado, func.sum(consolidated.columns.acciones), func.sum(consolidated.columns.monto)).\
group_by(consolidated.columns.estado).all()
query1
```

```python
query2 = session.query(fonden.columns.estado, func.count(fonden.columns.evento)).\
group_by(fonden.columns.estado).all()
query2
```

```python
df2 = pd.DataFrame({
"estado": [x[0] for x in query1],
"acciones": [x[1] for x in query1],
"monto": [x[2] for x in query1]
})
```

```python
df3 = pd.DataFrame({
"estado": [x[0] for x in query2],
"desastres": [x[1] for x in query2]
})
```
